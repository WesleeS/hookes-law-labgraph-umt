```python
#Physics plotting and least-squares fitting library
#to be imported 
#March 27, 2018
#Jaylene

import numpy as np
import matplotlib.pyplot as plt
import math

#define plotting and linear least squares fitting as functions

#goodness of fit parameter
def LLSFD2(x,y,dy,b,m):
    return sum(((y-b-m*x)/dy)**2)
                      
def Delta(x, dy):
    return (sum(1/dy**2))*(sum(x**2/dy**2))-(sum(x/dy**2))**2


def plotfit(x,y,dy,title,xlabel,ylabel,slopeunit):
    b,m=np.polynomial.polynomial.polyfit(x,y,1,w=dy)
    fit = b+m*x
    N = LLSFD2(x,y,dy,b,m)    
    D=Delta(x, dy)
    dm = math.sqrt(1/D*sum(1/dy**2)) #error in slope
    #db = math.sqrt(1/D*sum(x**2/dy**2)) #error in intercept
    #Commented out this line as... I don't think db is ever used?

    plt.scatter(x, y, color='red', marker='o')
    #Plot least squares fit line
    plt.plot(x, fit, color='green', linestyle='--')

    #Plot y error bars - do not need to modify
    plt.errorbar(x, y, yerr=dy, xerr=None, fmt="none")

    #Write x and y axes labels, title.  Modify these!
    plt.xlabel(xlabel) #x-axis label
    plt.ylabel(ylabel) #y-axis label
    plt.title(title) #plot title
    plt.show()
    
    print (f'''
    Slope = {m:.2f} {slopeunit}
    Error in slope = {dm:.2f} {slopeunit}
    Goodness of fit = {N:.2f}
    ''')

hungDisplacement = np.array([0,.03,.12,.185,.265,.350,.415])
hungMass = np.array([.05,.550,1.050,1.550,2.050,2.550,3.050])
restoringForce = hungMass*9.8

errm = np.array([1,1,1,1,1,1,1])/1000*9.8


plotfit(hungDisplacement,restoringForce,errm,'Steel(?) Spring','x Displacement(m)','Force of Gravity(N)','N/m')
    
### I'm assuming this was intentional for having the code not run, as a sort of 
### coding fitness test, though I might suggest having a guide present in the box
### folder with of how to do it

### EDIT: I realized yeah theres a demo in box haha, though it wouldn't be a bad
### idea to replace that with a step by step example of how to do it. 
### Not to say it's a bad thing to have it how it is currently, and I have nowhere
### close to your experience as a teacher, AND it's not like this is a class to teach
### python for new coders, but I think a guide of how to do it would be more cohesive
### than an example of the finished product as people are likely to just slot that 
### into a neighboring cell, add their data, and call it a day there.
```
